## Plot the Riemann function for a power x, after summing the first 'n' terms

## R(x, n) = Sum_{i=1}^{i=n} 1.0 / pow(i,x)

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [12, 8]

n = 1000

def riemann(p):
	y = 1.0 # just to start, this value is not used
	R = 0.0
	for i in range(1, n):
		y = 1.0 / pow(i, p)
		R += y
	return R;

x_low = 1.0
x_high = 5.0
num_steps = 30
xs = [x_low + i * (x_high - x_low) / num_steps for i in range(0, num_steps)]
Rs = [riemann(x) for x in xs]

print("Number of terms in sum = " + str(n))
plt.scatter(xs, Rs, lw=2, label = 'Riemann(x)')
plt.xlabel('x')
plt.legend(loc = 1)
plt.show()
```

## Plot the assymptotic behaviour of the Riemann function for a power p, as a function of the number of terms in the sim, until the step becomes smaller than the sum so far times 'step'.

## R(p) = Sum_{i=1}^{i=inf} 1.0 / pow(i,p)

```python
step = 1.0 / 1.e6

def riemann_to_plot(ns, ys, Rs, p):
	n = 1.0
	y = 1.0 # just to start, this value is not used
	R = 0.0
	while y > R * step:
		ns.append(n)
		y = 1.0 / pow(n, p)
		ys.append(y)
		R += y
		Rs.append(R)
		n += 1.0
	return R;

ns = []
ys = []
Rs = []

#Sum_{n=1}^inf(1/n^p)
p = 1.5
val = riemann_to_plot(ns, ys, Rs, p)
print (val)

plt_Rs, = plt.plot(ns, Rs, lw=2, color = 'red', label = 'Riemann(x) with n terms')
plt_ys, = plt.plot(ns, ys, lw=2, color = 'blue', label = 'n-th increment')
plt.xscale('log')
plt.xlabel('n')
plt.legend(loc = 4)
plt.show()
```
